1. Importar librerías

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

2. Cargar dataset desde internet

```python
# Usaremos un dataset clásico: Tips de propinas (Seaborn)
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv"
df = pd.read_csv(url)
```

3. Exploración inicial

```python
print("Primeras filas:")
print(df.head())
```

```python
print("\nInformación general:")
print(df.info())
```

```python
print("\nEstadísticas descriptivas:")
print(df.describe())
```

4. Detección de valores faltantes

```python
print("\nValores faltantes por columna:")
print(df.isnull().sum())
```

5. EDA gráfica

```python
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.hist(df['total_bill'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribución de Total Bill')
plt.xlabel('Total Bill')
plt.ylabel('Frecuencia')
plt.subplot(1, 2, 2)
sns.boxplot(x='day', y='total_bill', data=df, palette='pastel')
plt.title('Total Bill por Día')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6, 5))
sns.scatterplot(x='total_bill', y='tip', data=df, hue='time')
plt.title('Relación entre Total Bill y Tip')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.show()
```
